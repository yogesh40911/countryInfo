```python
from countryinfo import CountryInfo
name='India'
country =CountryInfo(name)
aboutName = country.alt_spellings()
print(aboutName)

capital=country.capital()
print(capital)

timeZone=country.timezones()
print(timeZone)

language=country.languages()
print(language)

translation=country.translations()
print(translation)

province=country.provinces()
print(province)

border=country.borders()
print(border)
```
